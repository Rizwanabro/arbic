```python
from transformers import AutoTokenizer
from transformers import AutoModel
from arabert.preprocess import ArabertPreprocessor
# model_name = "aubmindlab/bert-base-arabertv2"
model_name="aubmindlab/bert-base-arabertv01"
arabert_prep = ArabertPreprocessor(model_name=model_name)
tokenizer = AutoTokenizer.from_pretrained(model_name)
model = AutoModel.from_pretrained(model_name)
```

```python
train_text= train_dataset['Text']
train_term= train_dataset['Term']
train_polarity= train_dataset['Polarity']

test_text= test_dataset['Text']
test_term= test_dataset['Term']
test_polarity= test_dataset['Polarity']


```

```python
train_text = train_text.apply(arabert_prep.preprocess)
train_term = train_term.apply(arabert_prep.preprocess)
```

```python
train_text_tokenized = []
train_term_tokenized = []

def text_tokenization(text):
  arr=[]
  for i in text:
    inputs = tokenizer.encode_plus(i, return_tensors='pt',max_length=3)
    arr.append(inputs)
  return arr

train_text_tokenized = text_tokenization(train_text)
train_term_tokenized = text_tokenization(train_term)
```

```python
train_text_outputs = []
train_term_outputs = []

def to_model(text):
  arr=[]
  for i in text:
   with torch.no_grad():
    outputs = model(**i)
    arr.append(outputs)
  return arr

train_text_outputs = to_model(train_text_tokenized)
train_term_outputs = to_model(train_term_tokenized)
```

```python
train_text_embeddings = []
train_term_embeddings = []

def embedding(text):
  arr=[]
  for i in text:
   embeddings = i['last_hidden_state']
   arr.append(embeddings)
  return arr

train_text_embeddings = embedding(train_text_outputs)
train_term_embeddings = embedding(train_term_outputs)
```

```python
#test dataset embedding
test_text = test_text.apply(arabert_prep.preprocess)
test_term = test_term.apply(arabert_prep.preprocess)
```

```python
test_text_tokenized = []
test_term_tokenized = []

def text_tokenization(text):
  arr=[]
  for i in text:
   inputs = tokenizer.encode_plus(i, return_tensors='pt',max_length=3)
   arr.append(inputs)
  return arr

test_text_tokenized = text_tokenization(test_text)
test_term_tokenized = text_tokenization(test_term)
```

```python
test_text_outputs = []
test_term_outputs = []

def to_model(text):
  arr=[]
  for i in text:
   with torch.no_grad():
     outputs = model(**i)
     arr.append(outputs)
  return arr

test_text_outputs = to_model(test_text_tokenized)
test_term_outputs = to_model(test_term_tokenized)
```

```python
test_text_embeddings = []
test_term_embeddings = []

def embedding(text):
  arr=[]
  for i in text:
   embeddings = i['last_hidden_state']
   arr.append(embeddings)
  return arr

test_text_embeddings = embedding(test_text_outputs)
test_term_embeddings = embedding(test_term_outputs)
```

```python
from keras.utils import to_categorical

def encode_sentimens(polarity):
  if polarity=='negative':
   return 0
  elif polarity=='positive':
   return 1
  elif polarity=='neutral':
   return 2
  else:
   return 2

train_polarity=train_polarity.apply(encode_sentimens)
test_polarity=test_polarity.apply(encode_sentimens)

train_polarity = to_categorical(train_polarity, num_classes=3)
test_polarity = to_categorical(test_polarity, num_classes=3)
```

```python
!pip install keras-self-attention
```

```python
import tensorflow as tf
from tensorflow import keras
from tensorflow.keras.layers import Dense, GRU, concatenate, Flatten
import numpy as np
from keras_self_attention import SeqSelfAttention






train_text_outputs = to_model(train_text_tokenized)
train_term_outputs = to_model(train_term_tokenized)
```

```python
train_text_outputs[0]
```

```python

# Text embeddings model
model_text = tf.keras.models.Sequential()
model_text.add(GRU(100, return_sequences=True, input_shape=(3, 768)))
attention = SeqSelfAttention(attention_activation='sigmoid')
model_text.add(attention)
model_text.add(Flatten())
# Term embeddings model
model_term = tf.keras.models.Sequential()
model_term.add(GRU(100, return_sequences=True, input_shape=(3, 768)))
attention = SeqSelfAttention(attention_activation='sigmoid')
model_term.add(attention)
model_term.add(Flatten())
# Concatenate the outputs before the final classification layer
concatenated = concatenate([model_text.output, model_term.output])
reshape = keras.layers.Reshape((600,))(concatenated)
output_layer = Dense(3, activation='softmax')(reshape)
# Build the final model
model = keras.Model(inputs=[model_text.input, model_term.input], outputs=output_layer)
# Compile and train the model
model.compile(loss='categorical_crossentropy',optimizer='rmsprop', metrics=['accuracy'] )
```

```python
train_text_embeddings = np.stack([tensor.squeeze().numpy() for tensor in train_text_embeddings], axis=0)
train_term_embeddings = np.stack([tensor.squeeze().numpy() for tensor in train_term_embeddings], axis=0)
test_text_embeddings = np.stack([tensor.squeeze().numpy() for tensor in test_text_embeddings], axis=0)
test_term_embeddings = np.stack([tensor.squeeze().numpy() for tensor in test_term_embeddings], axis=0)
```

```python

model.fit([train_text_embeddings, train_term_embeddings], train_polarity, epochs=15, batch_size=256, validation_data=([test_text_embeddings, test_term_embeddings], test_polarity), validation_steps=1)
```

```python
model.evaluate([test_text_embeddings, test_term_embeddings], test_polarity)
```
